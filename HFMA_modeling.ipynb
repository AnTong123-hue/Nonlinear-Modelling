```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize parameter
# Helical coil 
d_oc = 1.49  # mm
d_ic = 1.10  # mm
k_c  = 0.011 # N/mm
# Microtubule
d_ot = 1.19  # mm
d_it = 0.64  # mm
E    = 1.648 # MPa
l_it = 50    # mm
# Microtubule elongation after insertion parameter
alpha = 0.9
```

```python
# Diameter of the microtubule after insertion 
d_o = d_ic
d_i = np.sqrt(d_o ** 2 - alpha * (d_ot ** 2 - d_it ** 2))
print (f"Diameter of the microtubule after insertion: {d_i:.2f} mm")
```

```python
# Elongation of microtubule
# random between 0 and 1.5
epsilon = np.random.rand(100) * 1.5
# Presure applied to the microtubule
P = ( alpha * E * np.pi /4 * (d_ot ** 2 - d_it ** 2) * (1 - 1/(1 + epsilon))  + k_c * epsilon * l_it ) / (np.pi / 4 * d_o ** 2 - alpha* np.pi / 4 * (d_ot ** 2 - d_it ** 2) * (1 - 1/(1 + epsilon)) )
```

```python
plt.scatter(epsilon, P)
plt.xlabel('Elongation (epsilon)')
plt.ylabel('Pressure (P)')
plt.title('Pressure vs Elongation of Microtubule')
plt.grid() 
plt.show()
```

```python
# Force applied to the microtubule
F = ( alpha * E * np.pi /4 * (d_ot ** 2 - d_it ** 2) * (1 - 1/(1 + epsilon))  + k_c * epsilon * l_it )
plt.scatter(epsilon, F, color='orange', marker='o', edgecolor='red')
plt.xlabel('Elongation (epsilon)')
plt.ylabel('Force (F)')
plt.title('Force vs Elongation of Microtubule')
plt.axis([0, 1.5, 0, 2.5])
plt.grid()
plt.show()
```
